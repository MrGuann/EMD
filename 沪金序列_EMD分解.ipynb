```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data = pd.read_excel('沪金指数_同花顺.xlsx')
close = data.close
date = data.datetime
```

```python
list_datetime = []
for i in range(len(date)):
    d = date[i].split(',')[0]
    list_datetime.append(d)
    #print(date[i])
```

```python
plt.plot(close)
```

```python
from PyEMD import EMD
```

```python
t = np.linspace(0,len(date)-1,len(date))
s = np.array(close)
```

```python
emd = EMD()
imfs = emd.emd(s,t)
```

```python
# Plot results
plt.figure(figsize=(12,15))
plt.subplot(7,2,1)
plt.plot(t, s, 'r')
plt.title("Input signal")

for n, imf in enumerate(imfs):
    plt.subplot(7,2,n+2)
    plt.plot(t, imf, 'g')
    plt.ylabel("IMF %i" %(n+1))

plt.tight_layout()
plt.savefig('simple_example')
plt.show()
```

```python
from PyEMD import EEMD
```

```python
eemd = EEMD()
eIMFs = eemd.eemd(s, t)
```

```python

nIMFs = eIMFs.shape[0]

# Plot results
plt.figure(figsize=(12,15))
plt.subplot((nIMFs//2)+1, 2, 1)
plt.plot(t, s, 'r')
plt.title("Input signal")

for n in range(nIMFs):
    plt.subplot((nIMFs//2)+1, 2, n+2)
    plt.plot(t, eIMFs[n], 'g')
    plt.ylabel("eIMF %i" %(n+1))
    plt.locator_params(axis='y', nbins=5)


plt.tight_layout()
plt.savefig('eemd', dpi=120)
plt.show()
```
